```python
high_points_y=np.array(SVC_scaled_CNN_C_gamma_result[SVC_scaled_CNN_C_gamma_result['f1_score']>0.83]['gamma_exp'])
high_points_x=np.array(SVC_scaled_CNN_C_gamma_result[SVC_scaled_CNN_C_gamma_result['f1_score']>0.83]['C_exp'])
```

```python
plt.title("SVM f1 score (C and gamma adjustment)")
X,Y = np.meshgrid([-5,-3,-1,1,3,5,7,9,11,13,15],[-15,-13,-11,-9,-7,-5,-3,-1,1,3])
Height = np.array(SVC_scaled_CNN_C_gamma_result['f1_score']).reshape(10,11)
plt.contourf(X, Y, Height, 10, alpha = 0.6, cmap = plt.cm.hot)
C = plt.contour(X, Y, Height, 10, colors = 'black', linewidth = 0.5)
plt.clabel(C, inline = True, fontsize = 10)
plt.scatter(high_points_x,high_points_y,label = "points got high f1_score")
plt.xlabel('C_exp')
plt.ylabel('gamma_exp')
plt.legend()
plt.show()
```

```python
plt.title("SVM BER (C and gamma adjustment)")
X,Y = np.meshgrid([-5,-3,-1,1,3,5,7,9,11,13,15],[-15,-13,-11,-9,-7,-5,-3,-1,1,3])
Height = np.array(SVC_scaled_CNN_C_gamma_result['BER']).reshape(10,11)
plt.contourf(X, Y, Height, 10, alpha = 0.6, cmap = plt.cm.hot)
C = plt.contour(X, Y, Height, 10, colors = 'black', linewidth = 0.5)
plt.clabel(C, inline = True, fontsize = 10)
plt.scatter(high_points_x,high_points_y,label = "points got high f1_score")
plt.xlabel('C_exp')
plt.ylabel('gamma_exp')
plt.legend()
plt.show()
```
